```python
files=os.listdir("data")
max_d=0
min_d=99999
gps_dict={}
for f in files:
    df=pd.read_csv("data/"+f,index_col=None, header=None)
    # df = df.reset_index(level=0)
    gps=[]
    for index, row in df.iterrows(): 
        gps.append(row.to_list())
        dis= np.linalg.norm(row.to_list())
        if dis>max_d:
            max_d=dis
        if dis<min_d:
            min_d=dis
    
    gps_dict[f]=gps
```

splitting into m regions

```python
def green(p1,p2):
    return np.dot(np.array(p1),np.array(p2))
```

```python
def regions(GPS,m):
    distances = [np.linalg.norm(v) for v in GPS]
    delta = (max_d-min_d) / m
    gps_reg=[[] for i in range(m)]
    for d in range(len(distances)):
        gps_reg[int((distances[d] - min_d) // delta) ].append(GPS[d])

    hist=[]
    for i in range(m):
        h1=[];
        for j in range(i+1,m):
            for p1 in gps_reg[i]:
                for p2 in gps_reg[j]:
                    #print(p1,p2)
                    h1.append(green(p1,p2))

            print(h1)

            plt.hist(h1)
            plt.show()



```

```python
regions(gps_dict["100gps.csv"],3)
```
